```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv("student_performance.csv")
```

```python
#  Line Plot (Overall% per student)
plt.figure(figsize=(8,5))
plt.plot(df["Name"], df["Overall"], marker="o")
plt.title("Overall Percentage per Student")
plt.xticks(rotation=90)
plt.ylabel("Percentage")
plt.show()
```

```python
# Bar Chart (Subject-wise scores)
df.plot(x="Name", y=["Maths", "Science", "English"], kind="bar", figsize=(10,5))
plt.title("Marks in Different Subjects")
plt.xticks(rotation=90)
plt.show()
```

```python
# Barplot - average marks in each subject
plt.figure(figsize=(8,5))
sns.barplot(data=df[["Maths", "Science", "English"]])
plt.title("Average Marks by Subject")
plt.show()
```

```python

# Boxplot - distribution of marks
plt.figure(figsize=(8,5))
sns.boxplot(data=df[["Maths", "Science", "English"]])
plt.title("Distribution of Marks")
plt.show()
```

```python
# Countplot - Grade distribution
plt.figure(figsize=(6,4))
sns.countplot(x="Grade", data=df, palette="Set2")
plt.title("Grade Distribution")
plt.show()
```

```python
import plotly.express as px

# Line chart (interactive)
fig = px.line(df, x="Name", y="Overall", title="Interactive Overall% per Student")
fig.show()

# Bar chart
fig = px.bar(df, x="Name", y=["Maths", "Science", "English"], barmode="group",
             title="Student Subject Marks")
fig.show()

# Pie chart (Grades)
fig = px.pie(df, names="Grade", title="Grade Distribution")
fig.show()
```
