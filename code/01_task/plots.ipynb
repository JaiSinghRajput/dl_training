```python
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.model_selection import train_test_split
import seaborn as sns
data = pd.read_csv("student_scores.csv")
```

```python
# 1. Print first 10 rows
print("First 10 rows of the dataset:")
display(data.head(10))
```

```python

# 2. Display shape (rows, columns)
print("\nShape of dataset (rows, columns):", data.shape)
```

```python

# 3. Descriptive statistics
print("\n--- Statistics ---")
print("Max Hours:", data["Hours"].max())
print("Min Hours:", data["Hours"].min())
print("Mean Hours:", data["Hours"].mean())
print("Median Hours:", data["Hours"].median())

print("\nMax Scores:", data["Scores"].max())
print("Min Scores:", data["Scores"].min())
print("Mean Scores:", data["Scores"].mean())
print("Median Scores:", data["Scores"].median())
```

```python

# --- Train-Test Split ---
train_set, test_set = train_test_split(
    data, test_size=0.3, random_state=42
)

print("\n--- Train-Test Split ---")
print("Training set rows:", train_set.shape[0])
print("Testing set rows:", test_set.shape[0])
```

```python

# --- Visualization ---
plt.figure(figsize=(8,6))

plt.scatter(train_set["Hours"], train_set["Scores"], 
            color="blue", label="Training Data", s=60)

plt.scatter(test_set["Hours"], test_set["Scores"], 
            color="orange", label="Testing Data", s=60)

plt.xlabel("Study Hours", fontsize=12)
plt.ylabel("Marks", fontsize=12)
plt.title("Study Hours vs Marks", fontsize=14)
plt.legend()
plt.grid(True)
plt.show()
```

```python
plt.figure(figsize=(12,5))

plt.subplot(1,2,1)
plt.hist(data["Hours"], bins=6, color="skyblue", edgecolor="black")
plt.xlabel("Study Hours")
plt.ylabel("Frequency")
plt.title("Distribution of Study Hours")

plt.subplot(1,2,2)
plt.hist(data["Scores"], bins=6, color="salmon", edgecolor="black")
plt.xlabel("Scores")
plt.ylabel("Frequency")
plt.title("Distribution of Scores")

plt.tight_layout()
plt.show()
```

```python
sorted_data = data.sort_values(by="Hours")
plt.figure(figsize=(8,6))
plt.plot(sorted_data["Hours"], sorted_data["Scores"], 
         marker="o", linestyle="-", color="green")
plt.xlabel("Study Hours")
plt.ylabel("Marks")
plt.title("Trend of Scores vs Study Hours")
plt.grid(True)
plt.show()
```

```python
plt.figure(figsize=(5,4))
sns.heatmap(data.corr(), annot=True, cmap="coolwarm", fmt=".2f")
plt.title("Correlation Heatmap")
plt.show()
```
